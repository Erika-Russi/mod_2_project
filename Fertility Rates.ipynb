```python
import pandas as pd
import csv
fertility=pd.read_csv('Fertility.csv')
fertility.head()
```

```python
density=pd.read_csv('Density.csv')
density
```

```python
# frames = [density, fertility] 
density_vs_fertility = density.merge(fertility, on='Country Name', how='left')
density_vs_fertility
```

```python
density_vs_fertility = density_vs_fertility.dropna(subset = ['2016_x'])
```

```python
density_vs_fertility = density_vs_fertility.dropna(subset = ['2016_y'])
```

```python
density_vs_fertility
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# %matplotlib inline
# plt.rcParams['figure.figsize'] = (8, 6)
# plt.rcParams['font.size'] = 14
```

```python
# Pandas scatter plot
plt.scatter(density_vs_fertility['2016_x'], density_vs_fertility['2016_y'])
# density_vs_fertility.plot(kind='scatter', x='2016_x', y='2016_y', alpha=0.2)
```

```python
#outliers above are ruining visibility
dvf = density_vs_fertility[density_vs_fertility.loc[:,'2016_x'] < 1000]
```

```python
plt.scatter(dvf['2016_x'], dvf['2016_y'])
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = (8, 6)
plt.rcParams['font.size'] = 14
dvf.plot(kind='scatter', x='2016_x', y='2016_y', alpha=0.2)
```

```python
#without outliers
sns.lmplot(x='2016_x', y='2016_y', data = dvf, aspect= 1.5, scatter_kws={'alpha':0.05})
```

```python
#with outliers
sns.lmplot(x='2016_x', y='2016_y', data = density_vs_fertility, aspect= 1.5, scatter_kws={'alpha':0.05})
```
